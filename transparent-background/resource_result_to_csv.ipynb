```python
import pandas as pd
```

```python
fg = pd.read_json("./results/False_cuda_monitoring.json")
fg["is_jit"] = False
fg["is_gpu"] = True
fg = fg[fg.columns[::-1]]

tg = pd.read_json("./results/True_cuda_monitoring.json")
tg["is_jit"] = True
tg["is_gpu"] = True
tg = tg[tg.columns[::-1]]

fc = pd.read_json("./results/False_cpu_monitoring.json")
fc["is_jit"] = False
fc["is_gpu"] = False
fc = fc[fc.columns[::-1]]

tc = pd.read_json("./results/True_cpu_monitoring.json")
tc["is_jit"] = True
tc["is_gpu"] = False
tc = tc[tc.columns[::-1]]

df_list = [fg, tg, fc, tc]
```

```python
new_df = pd.concat(df_list).reset_index(drop=True)
```

```python
new_df= new_df.rename(columns=\
                    {"is_gpu":"is_gpu", "is_jit":"is_jit", "scale":"scale", "params":"params", "flops":"flops", \
                    "max_time":"max_time/sec", "avg_time":"avg_time/sec",\
                    "max_gpu":"max_gpu/MB", "avg_gpu":"avg_gpu/MB",\
                    "max_ram":"max_ram/MB", "avg_ram":"avg_ram/MB",\
                    "max_cpu":"max_cpu/%", "avg_cpu":"avg_cpu/%"})
```

```python
new_df
```

```python
def replace_unit(text):
    if isinstance(text, str):
        if "M" in text:
            text = text.replace("M","")
            return float(text) * 1000000
        elif "G" in text:
            text = text.replace("G","")
            return float(text) * 10**9
    elif isinstance(text, (float, int)):
        if pd.isna(text):
            return -1
        else:
            return text
    else:
        text
```

```python
new_df = new_df.applymap(replace_unit)
```

```python
new_df["scale"] = new_df["scale"].apply(lambda x: str(x))
```
